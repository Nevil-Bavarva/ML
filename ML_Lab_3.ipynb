1_NB_Classifier_weather

```python
from sklearn import preprocessing
from sklearn.naive_bayes import GaussianNB , MultinomialNB
```

```python
weather =['Sunny', 'Sunny', 'Overcast', 'Rainy', 'Rainy','Rainy', 'Overcast',
'Sunny', 'Sunny', 'Rainy', 'Sunny', 'Overcast', 'Overcast', 'Rainy']

temp = ['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild',
'Cool','Mild','Mild','Mild','Hot','Mild']

play=['No','No','Yes','Yes','Yes','No','Yes','No','Yes',
'Yes','Yes','Yes','Yes','No']
```

```python
le = preprocessing.LabelEncoder()

weather_encoded = le.fit_transform(weather)

print("weather:" , weather_encoded)
```

```python
temp_encoded = le.fit_transform(temp)
label = le.fit_transform(play)

print("Temperature: ",temp_encoded)
print("Play:", label)
```

```python
features = tuple(zip(weather_encoded,temp_encoded))
print("Features:", features)
```

```python
model = GaussianNB()
model.fit(features,label)
```

```python
model = MultinomialNB()
model.fit(features,label)
```

```python
predicted = model.predict([[0,2]]) #overcast and hot
print("predicted value:", predicted)
```

```python
predicted = model.predict([[0,1]]) #overcast and mild
print("predicted value:", predicted)
```

```python
predicted = model.predict([[2,2]]) #sunny and mild
print("predicted value:", predicted)
```

2_NB_Classifier_Iris_2Classes

```python
import numpy as np
from sklearn import datasets , preprocessing
from sklearn.naive_bayes import GaussianNB
import pandas as pd


#iris = datasets.load_iris()
```

```python
data = pd.read_csv("/content/Dataset1.csv")
le = preprocessing.LabelEncoder()
```

```python
features = tuple(zip(le.fit_transform(data["Outlook"]),
                    le.fit_transform(data["Temp"]),
                    le.fit_transform(data["Humidity"]),
                      le.fit_transform(data["Wind"])))
print(features)
```

```python
play_encoded = le.fit_transform(data["Play"])
print(play_encoded)
```

```python
model = GaussianNB()
model.fit(features,play_encoded)
```

```python
for i in features:
  print(model.predict([i]))
```
